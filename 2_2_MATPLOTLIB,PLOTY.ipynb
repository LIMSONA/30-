# **Matplotlib**

- 데이터를 Chart, Plot으로 시각화해주는 라이브러리이다.
- 2003년 0.1 출시, 17년이된 패키지
- 가장 많이 사용되는 시각화 패키지이다.
- 기본 구성 : 그림(figure), 축(axes)
- % matplotlib inline : jupyter notebook 내에서 output을 보여준다.

##Basic



- alpha : 투명도
- logy : Y축 Log scaling
- kind : line, bar, barh, kde
- subplots : 여러개의 plot 그리기
- legend : subplot의 범례 지정
- xlim, ylim : X축과 Y축의 경계(axis로 한번에 그릴 수 있음)
- grid : 그리드 표현(True, False)
- title : 그래프의 제목 지정
- linewidth : 라인 넓이
- color : 색
- linestyle : 실선, 점선, 1점 쇄선 등
- marker : 마커 지정
- markerfacecolor : 마커 색
- markersize : 마커 크기

##label & legend

* label

  - 무슨 데이터인지 표시해주는 방법

* legend

  - 좌표축에 범례를 추가해주는 방법
  - loc : legend 위치
      - 'best' 0 (알아서 위치 찾음)
      - 'upper right' 1
      - 'upper left' 2
      - 'lower left' 3
      - 'lower right' 4
      - 'right' 5
      - 'center left' 6
      - 'center right' 7
      - 'lower center' 8
      - 'upper center' 9
      - 'center' 10

```python
#그래프 2개 동시에 그리기 
#왼쪽 아래에 범례 추가해서 넣기

x = np.linspace(0, 10, 100)
y = np.sin(x)
y_ = np.cos(x)

# 범례 추가하기
plt.plot(x, y, label='sin')
plt.plot(x, y_, '-o', label='cos')

# 4번 위치에 범례 넣기
plt.legend(loc=0)
```

##scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(1,10,20)
y=x**2

plt.scatter(x,y,color='r',alpha=0.5)
#alpha는 투명도
plt.show()
```

## histogram

- 도수분포표를 그래프로 나타낸 방법이다.
- 막대그래프는 계급 즉 가로를 생각하지 않고 세로의 높이로만 나타낸다. (출처 : 위키백과)
- 연속형 자료를 계급으로 나누어  계급별 도수를 막대로 나타낸다.

```python
x10= [np.random.randint(1,7) for i in range(10)] 
x1000= [np.random.randint(1,7) for i in range(1000)] 
x100000= [np.random.randint(1,7) for i in range(100000)] 
#1~6 주사위를 1000번 던진 경우
#10번보다 100번 / 100번보다 1000번 던질 떄 데이터 그래프높이가 평평해짐
#데이터 큰수의 법칙
```

```python
plt.hist(x10)
plt.show()
```

```python
plt.hist(x1000)
plt.show()
```

```python
plt.hist(x100000)
plt.show()
```

```python
#r그래프가 붙어서 보이니까 사이 간격을 bins로 조절 가능하다
plt.hist(x1000, bins=11)
plt.show()
```

```python
plt.hist(x1000, bins=8)
plt.show()
```

##pie chart

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels=labels)
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels=labels,shadow=True)
plt.show()
```

```python
labels = ['one', 'two', 'three']
size = [100, 20, 10]

plt.pie(size, labels=labels, shadow=True, autopct='%.2f%%') #자동비율 추가하기 소수점 둘째자리
plt.show()
```

##bar chart

- 범주형 자료의 분포를 파악한다.

```python
plt.bar(['one', 'two', 'three'], [10, 20, 30])
plt.show
```

```python
plt.barh(['one', 'two', 'three'], [10, 20, 30]) # barh는 가로방향 그래프
plt.show()
```

##선 색상과 스타일

- plt.plot(x, y, color='red') -> red, green, blue 등 색상 이름
- plt.plot(x, y, color='r') -> r, g, b, y, m(자홍), k(검정) 등 색상 이름
- plt.plot(x, y, color='0.2') -> 회색조(0-1사이 값)
- plt.plot(x, y, color='#ff0000') -> 16진수 색상 값

---

- plt.plot(x, y, linestyle='solid') -> 실선('-')
- plt.plot(x, y, linestyle='dashed') -> 파선('--')
- plt.plot(x, y, linestyle='dashdot') -> 1점 쇄선('-.')
- plt.plot(x, y, linestyle='dotted') -> 점선(':')

---

- plt.plot(x, y, '--r') -> 빨간색 파선

---

- '.' point marker
- 'o' circle marker(plt.plot(x, y_, '-o'))
- '^' triangle_up marker
- 's' square marker
- 'p' pentagon marker
- '*' star marker
- '+' plus marker
- 'x' x marker
